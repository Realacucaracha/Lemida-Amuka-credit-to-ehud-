### Define the Sigmoid Function and its Derivative
- Construct a function returning a sigmoid function:
$ \sigma(x) = \frac{1}{1 + e^{-x}} $
- Construct a function returning the derivative of a sigmoid function:
$ \frac{d\sigma(x)}{dx} = \sigma(x)(1 - \sigma(x)) $

### Initialize Weights
Build an array of three weights (3x1 array â€“ think why these dimensions!) and initialize their value randomly. (It is good practice to use weights with normal distribution of $ \mu = 0 $ and  $ \sigma = \frac{1}{3}  $ )

````python
def sigmo(x):
    return 1 / (1 + np.exp(-x))
def sigmo_der(x):
    return sigmo(x)*(1-sigmo(x))
````

````python
#BackPropagation:
#y_predict = np.array([[0],[0],[0],[0]])
bias = np.array([[0.2],[0.2],[0.2],[0.2]]) 
for i in range(3):
    weights[i][0] = np.random.normal(0,0.333) #mean=0, std = 1/3

#-------------------

y_predict = np.zeros((4,1))
for i in range(1000):

    #forward culculation:
    for q in range(len(X)):
        y_predict[q] = sigmo(np.dot(X[q],weights) + bias[q])
  

    #diffrence culc...(aka error)
    Ero = 0
    for w in range(len(y_predict)):
        Ero += (y_predict[w] - y_real[w]) 
    print("Numero:",i,"--->",Ero)

    #changing weights culc:
    for w in range(len(y_predict)):
        for i in range(len(weights)):

            diff = (y_predict[w] - y_real[w]) * y_predict[w] *(1 - y_predict[w]) * X[w][i]
            #print("diff is:",diff)
            weights[i] = weights[i] - learning_rat*diff
        
            bias_diff = (y_predict[w] - y_real[w]) * y_predict[w] *(1 - y_predict[w]) 
            bias[i] = bias[i] - learning_rat*bias_diff

    print("Now we:\n",weights,"Erorr->",Ero)

````
